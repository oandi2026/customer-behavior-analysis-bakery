```python
import pandas as pd

df = pd.read_csv("data/customer_survey.csv")


# 2. Menghitung total per jawaban menggunakan kolom 'count'
summary = df.groupby("answer")["count"].sum()

# 3. Menampilkan hasil summary
print(summary)
```

```python
import matplotlib.pyplot as plt

# 1. Tentukan urutan jawaban yang logis untuk kuesioner (Skala Likert)
urutan_jawaban = ['Strongly Agree', 'Agree', 'Don\'t Know', 'Disagree', 'Strongly Disagree']

# 2. Urutkan ulang data summary berdasarkan skala di atas
summary_sorted = summary.reindex(urutan_jawaban)

# 3. Buat grafik batang dengan warna yang menarik
summary_sorted.plot(kind='bar', color='#4a90e2', edgecolor='black')

# 4. Atur judul dan label grafik
plt.title("Distribusi Jawaban Kuesioner Pelanggan", fontsize=14, fontweight='bold', pad=15)
plt.xlabel("Pilihan Jawaban", fontsize=12)
plt.ylabel("Total Jumlah (Count)", fontsize=12)

# 5. Putar teks label bawah agar tidak saling menabrak dan mudah dibaca
plt.xticks(rotation=45, ha='right')

# 6. Sempurnakan tata letak agar tidak ada teks yang terpotong saat disimpan
plt.tight_layout()


# Tambahkan baris ini tepat sebelum plt.show()
plt.savefig('images/distribution_chart.png', dpi=300, bbox_inches='tight')
plt.show()


# 7. Tampilkan grafik
plt.show()
```
